## Rajkos blokk kérdései

```python
# 1. kérdés - Jelentkezes

respondent = []
potential_match = []
elso_kerdes = []

for i in range(len(dataframe)):
    for j in range(len(dataframe)):
        respondent.append(dataframe["Nev"][i])
        potential_match.append(dataframe["Nev"][j])
        if dataframe["Jelentkezes"][i] == dataframe["Jelentkezes"][j]:
            elso_kerdes.append(1)
        elif dataframe["Jelentkezes"][i] in [
            "Közösségre vágytál",
            "Szereted az öntevékenységet",
        ] and dataframe["Jelentkezes"][j] in [
            "Közösségre vágytál",
            "Szereted az öntevékenységet",
        ]:
            elso_kerdes.append(0.5)
        elif dataframe["Jelentkezes"][i] in [
            "Nem jött be az egyetem",
            "Valami pluszt akartál csinálni egyetem mellett",
            "Szakmai pezsgés",
        ] and dataframe["Jelentkezes"][j] in [
            "Nem jött be az egyetem",
            "Valami pluszt akartál csinálni egyetem mellett",
            "Szakmai pezsgés",
        ]:
            elso_kerdes.append(0.5)
        elif dataframe["Jelentkezes"][i] in [
            "Kapcsolati háló",
            "Lakhatás miatt",
            "Anno nem igazán tudtam, mi ez az egész, de szimpi volt!",
        ] and dataframe["Jelentkezes"][j] in [
            "Kapcsolati háló",
            "Lakhatás miatt",
            "Anno nem igazán tudtam, mi ez az egész, de szimpi volt!",
        ]:
            elso_kerdes.append(0.5)
        else:
            elso_kerdes.append(0.1)

szazalek["respondent"] = respondent
szazalek["potential_match"] = potential_match
szazalek["elso_kerdes"] = elso_kerdes
```

```python
# 2. kérdés - Kedvenc kurzus

uj_nev = {
    "PWT": ["PWT", "pwt", "Pwt"],
    "HÜK": ["Haladó üzleti elméletek", "HÜK"],
    "Monetáris politika": ["Monetáris politika", "Monpol", "monpol"],
    "Network Science": ["Network Science", "Network"],
    "Prog 1": ["Prog", "Prog 1"],
    "Prog 2": ["Prog2", "Prog 2"],
    "IR": [
        "IR",
        "ir",
        "Még nem tudom úf vagyok, de voltam international relations betekintőn, az tetszett:D",
    ],
    "Kischikán": ["kischikán"],
    "Pszichológia": ["Pszicho"],
    "Gedeon": ["Gedeon", "Társadalomelméletek a modernitásról"],
}

for new_value, old_values in uj_nev.items():
    dataframe["Kedvenc kurzus"].replace(old_values, new_value, inplace=True)

g_kar = [
    "Kritikai menedzsmentkutatás",
    "Vállalatelméletek",
    "HÜK",
    "Stratégia",
    "Zeneipar",
    "Kischikán",
]
k_kar = ["EWT", "Fejlődésgazdaságtan", "Makró", "Politikai gazdaságtan"]
t_kar = [
    "PWT",
    "Bevszoc",
    "Szociológia",
    "Feminizmus",
    "Gedeon",
    "IR",
    "Morálfiló",
    "Oktatáspolitika",
    "Politikai filozófia",
    "Vallásantropológia",
    "Gedeon",
]
adat = ["Network Science", "Machine Learning", "Prog 1", "Prog 2", "Ökó"]
penzugy = ["Monetáris politika", "PAK", "Vállalatértékelés"]
fless_egyeb = [
    "Biblia kurzus",
    "Bármelyik filozófiai",
    "Filmkurzus",
    "Kognitív Pszichológia",
    "Vidékfejlesztés",
    "Pszichológia",
]
nincs_match = ["B-kurzus", "Ehhez túl keveset vegeztem el.", "UF-ként még nincs"]

masodik_kerdes = []


def calculate_match_percentage(answer_i, answer_j):
    if answer_i == answer_j:
        return 1.0
    if (answer_i in nincs_match) or (answer_j in nincs_match):
        return 0.0
    if answer_i in g_kar and answer_j in g_kar:
        return 0.75
    if answer_i in k_kar and answer_j in k_kar:
        return 0.75
    if answer_i in t_kar and answer_j in t_kar:
        return 0.75
    if answer_i in adat and answer_j in adat:
        return 0.75
    if answer_i in penzugy and answer_j in penzugy:
        return 0.75
    if answer_i in fless_egyeb and answer_j in fless_egyeb:
        return 0.75
    return 0.0


for i in range(len(dataframe)):
    for j in range(len(dataframe)):
        match_percentage = calculate_match_percentage(
            dataframe["Kedvenc hely"][i], dataframe["Kedvenc hely"][j]
        )
        masodik_kerdes.append(match_percentage)

szazalek["masodik_kerdes"] = masodik_kerdes
```

```python
# 4. kérdés - Kedvenc hely

negyedik_kerdes = []


def calculate_match_percentage(answers_i, answers_j):
    set_i = set(answers_i.split(","))
    set_j = set(answers_j.split(","))
    intersection = set_i & set_j
    len_intersection = len(intersection)

    if len_intersection == 0:
        return 0
    elif len_intersection == 1:
        return 0.6
    elif len_intersection == 2:
        return 0.8
    else:
        return 1.0


for i in range(len(dataframe)):
    for j in range(len(dataframe)):
        match_percentage = calculate_match_percentage(
            dataframe["Kedvenc hely"][i], dataframe["Kedvenc hely"][j]
        )
        negyedik_kerdes.append(match_percentage)

szazalek["negyedik_kerdes"] = negyedik_kerdes
```

```python
# top5 match, végül nem ezt használom


def get_top_5_matches(df, respondent_name):

    # kiválasztom az adott nevű személyhez tartozó sorokat és rendezzük őket csökkenő sorrendben a százalékos értékek alapján
    matches = df[df["respondent"] == respondent_name].sort_values(
        by="final_percentage", ascending=False
    )

    # ha nincs találat az adott névre
    if matches.empty:
        return f"Nincs találat az adott névre: {respondent_name}"

    # top5 - százalékban adja meg a választ, két tizedesjegyig
    matches["final_percentage"] = matches["final_percentage"].apply(
        lambda x: f"{x*100:.2f}%"
    )
    top_5_matches = matches.head(5)

    return top_5_matches[["potential_match", "final_percentage"]]


# példa
respondent_name = "Mizsák Máté"
top_5 = get_top_5_matches(szazalek_filtered, respondent_name)
print(f"{respondent_name} legjobb 5 találata:\n{ top_5 }")
```

```python
# top 5 match


def get_top_5_matches(szazalek_filtered):

    # dictionary létrehozása a párokhoz és százalélokhoz
    top_5_matches = {}

    for respondent in szazalek_filtered["respondent"].unique():
        matches = szazalek_filtered[szazalek_filtered["respondent"] == respondent]
        matches_sorted = matches.sort_values(by="final_percentage", ascending=False)
        top_5 = matches_sorted.head(5)
        top_5_list = list(zip(top_5["potential_match"], top_5["final_percentage"]))
        top_5_matches[respondent] = top_5_list

    return top_5_matches


top_5_matches = get_top_5_matches(szazalek_filtered)
for respondent, matches in top_5_matches.items():
    print(f"{respondent}:")
    for match in matches:
        print(f"    {match[0]} - {match[1]}")
```

```python
# két ember közötti match százalék


def get_match_percentage(df, respondent_name, potential_match_name):

    # annak a sornak a vizsgálata, amiben a megadott nevek vannak
    match_row = df[
        (df["respondent"] == respondent_name)
        & (df["potential_match"] == potential_match_name)
    ]

    # ha nincs adat, ezt adja vissza
    if match_row.empty:
        return f"Nincs adat. :("

    # százalékban adja meg a választ, két tizedesjegyig
    szazalekos_forma = match_row.iloc[0]["final_percentage"] * 100
    return f"{szazalekos_forma:.2f}%"


# példa
respondent_name = "Mizsák Máté"
potential_match_name = "Barabás Dániel"

match_percentage = get_match_percentage(
    szazalek_filtered, respondent_name, potential_match_name
)
print(f"{respondent_name} + {potential_match_name} = {match_percentage}")
```

```python
# 2. lépés, hogy a korábban használt get_top_5_matches függvényt módosítom és kiterjesztem, hogy a top5 pár meglegyen hetero, meleg, összes kategóriákra

def get_top_5_matches(szazalek_filtered):
    # dictionary létrehozása a párokhoz és százalékokhoz
    top_5_matches = {}

    for respondent in szazalek_filtered["respondent"].unique():
        matches = szazalek_filtered[szazalek_filtered["respondent"] == respondent]

        # összesített top 5
        matches_sorted = matches.sort_values(by="final_percentage", ascending=False)
        top_5_all = matches_sorted.head(5)
        top_5_all_list = list(
            zip(
                top_5_all["potential_match"],
                top_5_all["final_percentage"],
                top_5_all["hetero_meleg"],
            )
        )

        # top 5 Hetero (0-ás értékű)
        matches_hetero = matches[matches["hetero_meleg"] == 0]
        matches_hetero_sorted = matches_hetero.sort_values(
            by="final_percentage", ascending=False
        )
        top_5_hetero = matches_hetero_sorted.head(5)
        top_5_hetero_list = list(
            zip(
                top_5_hetero["potential_match"],
                top_5_hetero["final_percentage"],
                top_5_hetero["hetero_meleg"],
            )
        )

        # top 5 Meleg (1-es értékű)
        matches_meleg = matches[matches["hetero_meleg"] == 1]
        matches_meleg_sorted = matches_meleg.sort_values(
            by="final_percentage", ascending=False
        )
        top_5_meleg = matches_meleg_sorted.head(5)
        top_5_meleg_list = list(
            zip(
                top_5_meleg["potential_match"],
                top_5_meleg["final_percentage"],
                top_5_meleg["hetero_meleg"],
            )
        )

        top_5_matches[respondent] = {
            "all": top_5_all_list,
            "hetero": top_5_hetero_list,
            "meleg": top_5_meleg_list,
        }

    return top_5_matches


top_5_matches = get_top_5_matches(szazalek_filtered)
for respondent, matches in top_5_matches.items():
    print(f"{respondent}:")
    print("    Összesített top 5:")
    for match in matches["all"]:
        print(f"        {match[0]} - {match[1]} - {match[2]}")
    print("    Top 5 Hetero:")
    for match in matches["hetero"]:
        print(f"        {match[0]} - {match[1]} - {match[2]}")
    print("    Top 5 Meleg:")
    for match in matches["meleg"]:
        print(f"        {match[0]} - {match[1]} - {match[2]}")
```

```python
# 3. lépés, hogy ezeket az adatokat is kimentem egy külön fájlba

# adatok mentése JSON formátumban 
with open("data2_v2_streamlit.json", "w", encoding="utf-8") as f:
    json.dump({"top_5_matches": top_5_matches}, f, ensure_ascii=False, indent=4)

with open("data2_v2_streamlit.json", "r", encoding="utf-8") as f:
    content = f.read()

print(content)
```
